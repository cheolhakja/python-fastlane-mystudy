```python
class MyModel:
    def __init__(self) -> None:
        self.a = torch.rand(1)
        self.b = torch.rand(1)

        #y = ax + b a,b  
        self.a = torch.nn.Parameter(self.a)
        self.b = torch.nn.Parameter(self.b)

    def forward(self, tmp):
        return self.a * tmp + self.b

```
