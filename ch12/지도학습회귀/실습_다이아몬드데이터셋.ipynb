```python
import torch
```

```python
class MyModel:
    def __init__(self) -> None:
        self.a = torch.rand(1)
        self.b = torch.rand(1)

        #y = ax + b에서 a,b를 파라미터로 등록한다
        self.a = torch.nn.Parameter(self.a)
        self.b = torch.nn.Parameter(self.b)

    def forward(self, tmp):
        return self.a * tmp + self.b

```

```python
# 데이터 준비
carat = torch.tensor(df["carat"], dtype=torch.float) 
price = torch.tensor(df["price"], dtype=torch.float) 

print(carat[:10])
print(len(carat))
print(price[:10])
print(len(price))
```
